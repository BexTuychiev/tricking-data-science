```python
import pandas as pd
import seaborn as sns

pd.options.plotting.backend = "plotly"

diamonds = sns.load_dataset("diamonds")
diamonds = diamonds.sample(10000)

fig = diamonds.plot(x="price", y="carat", kind="scatter")

fig.show()
```
