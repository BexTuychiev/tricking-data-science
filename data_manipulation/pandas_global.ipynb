## Setting Plotly as Pandas plotting backend

How to make Pandas shake hands w覺th Plotly and deeply 覺nsult Matplotl覺b.

Or 覺n other words, let's break up the dream duo:

![](../images/2022/6_june/pandas_plotly_backend.png)

```python
import pandas as pd
import seaborn as sns

diamonds = sns.load_dataset("diamonds")
diamonds = diamonds.sample(10000)

diamonds.plot(x="price", y="carat", kind="scatter");
```

```python
import pandas as pd
import seaborn as sns

pd.options.plotting.backend = "plotly"

diamonds = sns.load_dataset("diamonds")
diamonds = diamonds.sample(10000)

fig = diamonds.plot(x="price", y="carat", kind="scatter")

fig.show()
```
